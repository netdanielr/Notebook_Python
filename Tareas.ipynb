```python
def segmentacion_edad(edad):
    if (edad >=0):
        if edad >=50:
            segmentacion = 'anciano'
        elif(edad >=18):
            segmentacion = 'adulto'
        elif(edad >=13):
            segmentacion = 'adolecente'
        elif(edad >= 5):
            segmentacion = 'niño'
        else:
            segmentacion = 'bebe'
        
        print('la edad ingresada es de un: ',segmentacion) 
    else:                                 
        print("Por favor ingresar edad correcta")
```

```python
segmentacion_edad(5)
```

```python
def fib(num):
    fibonacci=[]
    x=0
    y=1
    for n in range(num):
        fibonacci.append(x+y)
        aux = x + y
        x = y
        y = aux
    
    c=fibonacci.index(num)

    return fibonacci[c-2:c]
```

```python
fib(3)
```

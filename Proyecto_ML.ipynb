```python
import datetime
date = datetime.date(int('2017'),int('12'),int('21'))
date2 = datetime.date(int('2017'),int('12'),int('24'))
x=date2 - date
print(x)
```

```python
def switch_mes(argument):
    argument = argument.lower()
    switcher = {
        "jan": 1,
        "feb": 2,
        "mar": 3,
        "apr": 4,
        "may": 5,
        "jun": 6,
        "jul": 7,
        "aug": 8,
        "sep": 9,
        "oct": 10,
        "nov": 11,
        "dec": 12
    }
    return  switcher.get(argument, argument)
```

```python
def switch_to_date(fecha_texto):
    fechaconvert = datetime.date(int(fecha_texto.split('-')[0]),
                             int(switch_mes(fecha_texto.split('-')[1])),
                             int(fecha_texto.split('-')[2]))+2000
    return  fechaconvert
```

```python
def switch_to_date2(fecha_texto):
    fechaconvert = datetime.date(int(fecha_texto.split('-')[2])+2000,
                             int(switch_mes(fecha_texto.split('-')[1])),
                             int(fecha_texto.split('-')[0]))
    return  fechaconvert
```

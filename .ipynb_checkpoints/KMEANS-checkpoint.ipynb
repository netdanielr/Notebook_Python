## kmeans

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 2, random_state = 1)
```

```python
kmeans.fit(df.drop("Private",axis=1))
```

```python
kmeans.cluster_centers_
```

## Evaluando el modelo de clustering

```python
df.head()
```

```python
def convertir_var_private(cluster):
    if cluster == "Yes":
        return 1
    else :
        return 0
        
```

```python
df["Cluster_real"] = df["Private"].apply(convertir_var_private)
```

```python
df["Cluster_asignado"] = kmeans.labels_
```

```python
df.head()
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(df["Cluster_real"] , kmeans.labels_))
```

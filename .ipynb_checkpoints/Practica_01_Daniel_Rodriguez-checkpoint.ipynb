# 3. PREPROCESAMIENTO DE DATOS

## 3.1 Tratamiento de Valores Nulos¶

```python
data2.isnull().sum()
```

```python
#### Imputar valores missing
data2["SEXO"].fillna(data2["SEXO"].mode()[0], inplace = True)
data2["FLAG_CASADO"].fillna(data2["FLAG_CASADO"].mode()[0], inplace = True)
data2["NRO_DEPENDIENTES"].fillna(data2["NRO_DEPENDIENTES"].mode()[0], inplace = True)
data2["FLAG_TRAB_INDEP"].fillna(data2["FLAG_TRAB_INDEP"].mode()[0], inplace = True)
data2["INGRESOS_COSOLICITANTE"].fillna(data2["INGRESOS_COSOLICITANTE"].min(), inplace = True)
data2["MONTO_PRESTAMO_MILES"].fillna(data2["MONTO_PRESTAMO_MILES"].min(), inplace = True)
data2["PLAZO_PRESTAMO_MESES"].fillna(data2["PLAZO_PRESTAMO_MESES"].mean(), inplace = True)
data2["FLAG_HISTORIAL_CREDITICIO"].fillna(data2["FLAG_HISTORIAL_CREDITICIO"].mode()[0], inplace = True)

#data2["ESTADOCIVIL"].fillna(data["ESTADOCIVIL"].mode()[0], inplace = True)
```

## 3.2 Conversion de Variables

```python
df = data2.copy()
```

```python
df.head(10)
```

```python
# SEXO-- Variable Dummy
dicc_sexo = {'Male': 1,
             'Female': 0}
df['SEXO'] = df['SEXO'].map(dicc_sexo)
```

```python
#  FLAG_CASADO -- Variable Dummy
dicc_flag_Casado = {'Yes': 1,
     'No': 0}
df['FLAG_CASADO'] = df['FLAG_CASADO'].map(dicc_flag_Casado)
```

```python
#  EDUCACION -- Variable Dummy
dicc_flag_Educa = {'Graduate': 1,
                     'Not Graduate': 0}
df['EDUCACION'] = df['EDUCACION'].map(dicc_flag_Educa)
```

```python
#  FLAG_TRAB_INDEP -- Variable Dummy
dicc_flag_TranInde = {'Yes': 1,
                     'No': 0}
df['FLAG_TRAB_INDEP'] = df['FLAG_TRAB_INDEP'].map(dicc_flag_TranInde)
```

```python
df
```

```python
#  FLAG_HISTORIAL_CREDITICIO -- Variable Dummy
dicc_flag_HisCred = {'Bueno': 1,
                     'Malo': 0}
df['FLAG_HISTORIAL_CREDITICIO'] = df['FLAG_HISTORIAL_CREDITICIO'].map(dicc_flag_HisCred)
```

```python
df_ec_dummies = pd.get_dummies(df["TIPO_ZONA"], prefix = 'TIPO_ZONA', drop_first = False)
```

```python
df_ec_dummies_dep = pd.get_dummies(df["NRO_DEPENDIENTES"], prefix = 'NRO_DEPENDIENTES', drop_first = False)
```

```python
df_ec_dummies
```

```python
del df["NRO_DEPENDIENTES"]
```

```python
del df["TIPO_ZONA"]
```

```python
df = pd.concat([df,df_ec_dummies], axis = 1)
```

```python
df = pd.concat([df,df_ec_dummies_dep], axis = 1)
```

```python
#FLAG_CRED_HIPO
dicc_FLAG_CRED_HIPO = {'Y': 1,
                     'N': 0}
df['FLAG_CRED_HIPO'] = df['FLAG_CRED_HIPO'].map(dicc_FLAG_CRED_HIPO)
```

```python
df
```

## 3.3 Preselección de Variables

```python
list(df.columns)
```

```python
features_to_model =['SEXO',
 'FLAG_CASADO',
 'EDUCACION',
 'FLAG_TRAB_INDEP',
 'INGRESOS_SOLICITANTE',
 'INGRESOS_COSOLICITANTE',
 'MONTO_PRESTAMO_MILES',
 'PLAZO_PRESTAMO_MESES',
 'FLAG_HISTORIAL_CREDITICIO',
 'TIPO_ZONA_Rural',
 'TIPO_ZONA_Semiurban',
 'TIPO_ZONA_Urban',
 'NRO_DEPENDIENTES_0',
 'NRO_DEPENDIENTES_1',
 'NRO_DEPENDIENTES_2',
 'NRO_DEPENDIENTES_3+']
```

```python
X_data = df[features_to_model]
X_data.head()
```

```python
y_data = df['FLAG_CRED_HIPO']
```

```python
# Separar Data de Training y data de Test
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, 
                                                    train_size = 0.60, 
                                                    random_state = 20)
```

# 4. MODELAMIENTO

## 4.1 ÁRBOL DE DECISIÓN

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Declarar el modelo a usar
model = DecisionTreeClassifier(max_depth = 3)
```

```python
# Entrenar (ajustar) el modelo
model = model.fit(X_train, y_train)
```

```python
pred_train = model.predict(X_train)
pred_test = model.predict(X_test)
```

## 4.1.2 Evaluación del Modelo

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
confusion_matrix(y_train, pred_train)
```

```python
accuracy_train = accuracy_score(y_train, pred_train)
print("El accuracy del modelo en data de training es:", accuracy_train)
```

```python

confusion_matrix(y_test, pred_test)
```

```python
accuracy_test = accuracy_score(y_test, pred_test)
print("El accuracy del modelo en data de testing es:", accuracy_test)
```
